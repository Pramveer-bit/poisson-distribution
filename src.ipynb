```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
placement_per_day= np.random.poisson(lam=4, size=4200)
```

```python
placement_per_day
```

```python
x_axis = np.unique(placement_per_day)
x_axis
```

```python
y_axis = list()
for x in x_axis:
    frequency = np.count_nonzero(placement_per_day == x)
    relative_frequency= frequency/placement_per_day.shape[0]
    y_axis.append(relative_frequency)
```

```python
y_axis
```

```python
plt.xlabel("number of students placed in one day")
plt.ylabel("probability of students placed per day")
plt.bar(x_axis, y_axis)
```

its a right skewed distribution

```python
weights =np.random.normal(loc=65,scale=5,size=280000000)
```

```python
weights
```

```python
min = np.min(weights)
max = np.max(weights)
```

```python
min 
```

```python
max 
```

```python
range_width= (max - min)/10
```

```python
range_width
```

```python
plt.xlabel("weights in kg")
plt.ylabel("frequency")
plt.title("frequency distribution")
axes = plt.hist(weights, bins=500)
```

```python
y_axis=axes[0]
x_axis = axes[1][1:]
```

```python
x_axis
```

```python
y_axis
```

```python
normalized_y_axis= y_axis/280000000
```

```python
normalized_y_axis
```

```python
plt.bar(x_axis, normalized_y_axis)
```

```python
z_score = (weights - np.mean(weights))/np.std(weights)
```

```python
z_score
```

```python
plt.xlabel("z-score")
plt.ylabel("frequency")
axes = plt.hist(z_score, bins=500)
```
